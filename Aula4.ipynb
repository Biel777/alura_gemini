```python
generation_config = {
    "candidate_count": 1,
    "temperature":0.5,
}
```

```python
safety_settings = {
    "HARASSMENT": "BLOCK_NONE",
    "HATE": "BLOCK_NONE",
    "SEXUAL": "BLOCK_NONE",
    "DANGEROUS": "BLOCK_NONE"
}
```

Inicionalizando o Modelo

```python
model = genai.GenerativeModel(model_name="gemini-1.0-pro",
                              generation_config=generation_config,
                              safety_settings=safety_settings)
```

```python
response = model.generate_content("Write a story about a big butterfly that became a banker to save her son.")
print(response.text)
```

```python
chat = model.start_chat(history=[])
```

```python
prompt = input("Esperando Prompt: ")

while prompt != "fim":
  response = chat.send_message(prompt)
  print("Resposta: ", response.text, "\n")
  prompt = input("Esperando prompt: ")
```
